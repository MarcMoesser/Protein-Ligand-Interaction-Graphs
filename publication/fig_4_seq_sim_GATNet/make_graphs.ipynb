```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import matplotlib.patches as mpatches
import matplotlib.ticker as plticker
import seaborn as sns
```

# Get data

```python
#save for plotting
input_folder = "data/"

GATNet_PC_df = pd.read_csv(input_folder+"GATNet_PC_ensemble_df.csv", index_col=0)
GATNet_RMSE_df = pd.read_csv(input_folder+"GATNet_RMSE_ensemble_df.csv", index_col=0)

#GATNet_PC_SD_df = pd.read_csv(target_folder+"GATNet_PC_SD.csv")
#GATNet_RMSE_SD_df = pd.read_csv(target_folder+"GATNet_RMSE_SD.csv")
```

```python
GATNet_PC_df.rename(columns={"PIG PC": "GATNet+PIG PC"}, inplace=True)
GATNet_RMSE_df.rename(columns={"PIG RMSE": "GATNet+PIG RMSE"}, inplace=True)
```

# Plot Results

```python
def plot_PC(PC_df, name):
    sns.set(font_scale = 1.5)
    sns.set_style("whitegrid", {"grid.color": ".7", "grid.linestyle": ":"})

    fig, axs = plt.subplots(figsize=(10, 10), dpi=300)

    axs = sns.lineplot(data=PC_df, linewidth = 3)

    axs.set_ylabel("Pearson correlation coefficient", fontsize = 20, weight="bold")
    axs.set_xlabel("Sequence identity cutoff", fontsize = 20, weight="bold")
    sns.move_legend(axs, "upper right")

    fig.tight_layout()
    fig.savefig(name+"_PC_performance.png", transparent=False, facecolor="w", dpi=300)
    #plt.close()

def plot_RMSE(RMSE_df, name):
    sns.set(font_scale = 1.5)
    sns.set_style("whitegrid", {"grid.color": ".7", "grid.linestyle": ":"})

    fig, ax2 = plt.subplots(figsize=(10, 10), dpi=300)

    ax2 = sns.lineplot(data=RMSE_df, linewidth = 3, palette=["red"])
    ax2.set_ylabel("Root mean square error (RMSE)", fontsize = 20, weight="bold")
    sns.move_legend(ax2, "upper center", bbox_to_anchor=(0.5, 0.95))
    #ax2.set_yticklabels([1.2,1.3,1.4,1,5,1.6,1.7], fontsize = 15, rotation = 0, weight="bold")
    plt.ylim(1.15, 1.75)
    ax2.set_xlabel("Sequence identity cutoff", fontsize = 20, weight="bold")
    sns.move_legend(ax2, "upper right")

    fig.tight_layout()
    fig.savefig(name+"_RMSE_performance.png", transparent=False, facecolor="w", dpi=300)
    #plt.close()
    
def plot_RMSE_PC_together(PC_df, RMSE_df, name):
    sns.set(font_scale = 1.5)
    sns.set_style("ticks", {"grid.color": ".7", "grid.linestyle": ":"})

    fig, axs = plt.subplots(figsize=(12, 10), dpi=300)

    axs = sns.lineplot(data=PC_df, linewidth = 3)

    axs.set_ylabel("Pearson correlation coefficient (PC)", fontsize = 20, weight="bold")
    axs.set_xlabel("Sequence identity cutoff", fontsize = 20, weight="bold")
    sns.move_legend(axs, "upper center", bbox_to_anchor=(0.481, 1))
    #plt.ylim(0.55, 0.9)
    #axs.set_yticklabels([0.6,0.65,0.7,0.75,0.8,0.85])

    ax2 = axs.twinx()
    ax2 = sns.lineplot(data=RMSE_df, linewidth = 3, palette=["red"])
    ax2.set_ylabel("Root-mean-square error (RMSE)", fontsize = 20, weight="bold")
    sns.move_legend(ax2, "upper center", bbox_to_anchor=(0.5, 0.95))
    #ax2.set_yticklabels([1.2,1.3,1.4,1,5,1.6,1.7], fontsize = 15, rotation = 0, weight="bold")
    plt.ylim(1.15, 1.75)

    fig.tight_layout()
    fig.savefig(name+"_PC_RMSE_performance.png", transparent=False, facecolor="w", dpi=300)
    #plt.close()
    
def plot_RMSE_and_PC(PC_df, RMSE_df, name):
    
    plot_PC(PC_df, name)
    plot_RMSE(RMSE_df, name)
    plot_RMSE_PC_together(PC_df, RMSE_df, name)
```
